```python
class Account_details:
    def __init__(self,bal,acc):
        self._balance=bal
        self._accountNo = acc
    def debit(self,amount):
        self._balance-=amount
        print("amount debited:",amount)
        #print("total amount:",self._balance)
    def credit(self,amount):
        self._balance+=amount
        print("amount credited:",amount)
     #   print("total amount:",self._balance)
    def get_balance(self):
        return self._balance
        
   
```

```python
acc1 = Account_details(10000,3862457)
print("-----ACCOUNT DETAILS-------")
print("Acount 1 account no:",acc1._accountNo)
print("Acount 1 balance:",acc1._balance)

acc1.debit(2000)
print("Total amount:",acc1.get_balance())
acc1.credit(3000)
print("Total amount:",acc1.get_balance())
    
```
