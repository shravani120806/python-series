```python
def fact(n):
    fact=1
    for i in range(1,n+1):
        fact*=i
    print(fact)

fact(3)
```

```python
def env_odd(n):
    if(n%2==0):
        print(n," is even")
    else:
        print(n," is odd")

env_odd(4)
env_odd(5)
```

```python
#recrusion
#n to 1
def show(n):
    if(n==0):
        return
    print(n,end=" ")
    show(n-1)

show(7)
```

```python
#sum of 1st n natural no
def sum_cal(n):
    #um=0
    if(n==0):
        return 0
    return sum_cal(n-1)+n
   
sum=sum_cal(4)
print(sum)
```
